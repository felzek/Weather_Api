<h1>Weather API Data Mining Results</h1>

<p><span style="font-weight:600;font-size:15px;">Observation 2:</span> Tempature gets higher around the equator.</p>
<p><span style="font-weight:600;font-size:15px;">Observation 3:</span> Latitude is correlated with latitude.</p>

```python
import requests
import json
import matplotlib.pyplot as plt
import requests as req
import pandas as pd
import json
import random
from config import apikey
from citipy import citipy
import openweathermapy.core as ow
```

```python

lat = []
lon = []
cities = set([])


while len(cities) < 1500:
    lati = (random.randrange(-90,90,1))
    long = (random.randrange(-180,180,1))
    city = citipy.nearest_city(lati,long)
    city = city.city_name
    lat.append(lati)
    lon.append(long)
    cities.add(city)

my_list = list(cities)
print(my_list)
```

```python
settings = {"units": "Imperial", "appid": apikey}

weatherz = []


for city in my_list:
    if len(weatherz) < 500:
        try:
            data = ow.get_current(city, **settings)
            weatherz.append(data)
           
        except:
            continue
    
print(len(weatherz))
```

```python
summary = ['name','sys.country',"coord.lat", "coord.lon",'main.temp','wind.speed','main.humidity','clouds.all']

info = [response(*summary) for response in weatherz]






column_names = ["name",'Country',"Latitude", "Longitude","Temperature",'Wind Speed (mph)','Humidity (%)','Cloudiness (%)']
weather_data = pd.DataFrame(info,columns=column_names)
weather_data
    
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(12,8))
plt.grid(True)
plt.scatter(weather_data['Temperature'],weather_data['Latitude'],color='r')
plt.title('Tempature VS Latitude in Cities in the World - 11/13/2017')
plt.xlabel('Latitude')
plt.ylabel('Tempature')


plt.savefig('Tempature vs Latitude in Cities in the World')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(12,8))
plt.grid(True)
plt.scatter(weather_data['Wind Speed (mph)'],weather_data['Latitude'],color='r')
plt.title('Wind Speed (mph) VS Latitude in Cities in the World - 11/13/2017')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.savefig('Wind Speed (mph) VS Latitude in Cities in the World')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(12,8))
plt.grid(True)
plt.scatter(weather_data['Cloudiness (%)'],weather_data['Latitude'],color='r')
plt.title('Cloudiness (%) VS Latitude in Cities in the World - 11/13/2017')
plt.xlabel('Cloudiness (%)')
plt.ylabel('Tempature')
plt.savefig('Cloudiness (%) VS Latitude in Cities in the World')
plt.show()
```
